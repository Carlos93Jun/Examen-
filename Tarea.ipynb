```python
mean = 39
desv_std = 15
x = 62
```

# Tarea
Sobre el laboratorio Aerolíneas:
¿Cuánta es la probabilidad de
que un cliente tenga menos de
62 años?
¿Cuál es el máximo de edad que
un cliente que tendrá con un 5%
de probabilidad?
Nota: Asuma que los datos siguen
dist. normal.

```python
# 1. Probabilidad de que X < 62
# Estandarizamos
z = (x - mean) / desv_std
round(z,2)
```

```python
# Respuesta: La probabilidad para el valor z de 1.53 es 94%, entonces la probabilidad de que un cliente tenga menos de 62 años es 94%
```

```python
# 2. Max de edad que un cliente tendrá con un 5% de probabilidad
# Emcontraremos el valor de x tal que el area de la izquierda de ese valor en la distribucion normal estandar sea del 95%
# el puntaje correspondiente a z con 95% es 1.64
Z = 1.64
X = mean + Z*desv_std
round(X)

# Respuesta: Por lo tanto con un 5% de probabilidad, un cliente tendra una edad maxima de aproximadamente 64 años
```
